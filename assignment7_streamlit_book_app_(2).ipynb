- When the user submits the form:

```python
# Write your code for Part 3 here

# TODO: Create a subheader for the form
# Example:
# st.subheader("Add a New Book")

# TODO: Create a form with title, author, genre, pages, read
# Example structure:
#
# with st.form("add_book_form"):
#     title = st.text_input("Title")
#     author = st.text_input("Author")
#     genre = st.text_input("Genre")  # or st.selectbox(...)
#     pages = st.number_input("Pages", min_value=1, step=1)
#     read = st.checkbox("Read")
#
#     submitted = st.form_submit_button("Add Book")
#
#     if submitted:
#         # validation and update logic here
#         pass

# TODO: When the form is submitted, validate the inputs
# - If title or author is empty: show st.error(...)
# - If pages <= 0: show st.error(...)

# TODO: If valid, append the new row to the DataFrame
#       and show a success message, e.g. st.success("...")

# TODO: Re-display the updated DataFrame so the user can see the new book
# Write your code for Part 3 here

import streamlit as st
import pandas as pd

# Section header
st.subheader("Add a New Book")

# Get the current DataFrame
books_df = st.session_state.get("books_df", None)

if books_df is None:
    st.warning("No book data found. Please make sure Part 1 has been executed.")
else:
    # --- Add Book Form ---
    with st.form("add_book_form"):
        title = st.text_input("Title")
        author = st.text_input("Author")
        genre = st.text_input("Genre")
        pages = st.number_input("Pages", min_value=1, step=1)
        read = st.checkbox("Read")

        submitted = st.form_submit_button("Add Book")

        if submitted:
            # --- Validation ---
            if title.strip() == "" or author.strip() == "":
                st.error("Title and Author cannot be empty.")
            elif pages <= 0:
                st.error("Pages must be greater than zero.")
            else:
                # Create new row
                new_book = pd.DataFrame([{
                    "Title": title,
                    "Author": author,
                    "Genre": genre,
                    "Pages": pages,
                    "Read": read
                }])

                # Append to session_state DataFrame
                st.session_state["books_df"] = pd.concat(
                    [st.session_state["books_df"], new_book],
                    ignore_index=True
                )

                st.success(f"Book '{title}' added successfully!")

    # Show updated DataFrame
    st.subheader("Updated Book Library")
    st.dataframe(st.session_state["books_df"])
```
